```python
import pandas as pd
```

```python
cep_censo = pd.read_csv('/content/drive/MyDrive/Data Science/Puc Rio DS/02_Sprint Eng Dados/MVP_Data Engineering/Data/CEP-Censo/census_code_cep_coordinates.csv')

cep_censo
```

```python
#Removendo o - das linhas de cep

cep_censo['POSTCODE'] = cep_censo['POSTCODE'].str.replace('-', '')
```

```python
cep_censo['POSTCODE'] = pd.to_numeric(cep_censo['POSTCODE'], errors= 'coerce')
```

```python
cep_censo
```

```python
# Abrindo base ceps de cidade de SP

cep_sp = pd.read_csv('/content/drive/MyDrive/Data Science/Puc Rio DS/02_Sprint Eng Dados/MVP_Data Engineering/Data/cep_sp.csv')

cep_sp
```

```python
# Abrindo base ceps de cidade de RJ

cep_rj = pd.read_csv('/content/drive/MyDrive/Data Science/Puc Rio DS/02_Sprint Eng Dados/MVP_Data Engineering/Data/cep_rj.csv')

cep_rj
```

```python
# concatenando ceps sp e rj

base_sp_rj = pd.concat([cep_sp, cep_rj])
```

```python
base_sp_rj
```

```python
# separando apenas os ceps de sp e rio da base cep-censo

filtered_cep = pd.merge(cep_censo, base_sp_rj, left_on= 'POSTCODE', right_on= 'Cep', how= 'inner')
```

```python
filtered_cep.drop(columns= ['POSTCODE', 'Numero'], inplace= True)
```

```python
filtered_cep
```

```python
# Specify the path where you want to save the CSV file
csv_file_path = '/content/drive/My Drive/cep_censo_filtrado.csv'

# Save the DataFrame to CSV
filtered_cep.to_csv(csv_file_path, index=False)
```
